**15시 데이터**

```python
import requests

url = "https://data.ex.co.kr/openapi/restinfo/restWeatherList?key=3313999031&type=json&sdate=20231101&stdHour=15"

r = requests.get(url)

data = r.json()
print(data)
```

**머신러닝 훈련**

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(rw_dfns, random_state=42)
print(len(train_set),len(test_set))
```

```python
x_train = train_set[['현재기온값']]
y_train = train_set[['시간대']]

print(x_train.shape, y_train.shape)
```

**선형 회귀 모델**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

**훈련된 모델 평가하기**

```python
x_train = train_set[['현재기온값']]
y_train = train_set[['시간대']]

lr.score(x_train, y_train)
```
