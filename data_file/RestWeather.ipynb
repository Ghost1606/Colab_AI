```python
import requests

url = "https://data.ex.co.kr/openapi/restinfo/restWeatherList?key=3313999031&type=json&sdate=20231101&stdHour=15"

r = requests.get(url)

data = r.json()
print(data)
```

```python
data
```

```python
rw = []
for d in data['list']:
  rw.append(d)

rw
```

```python
import pandas as pd
rw_df = pd.DataFrame(rw)

rw_df
```

```python
#file_dir = '/content/drive/MyDrive/Colab_Ai/Colab_AI_Git/Data_file'
rw_df.to_json('rest_weather.json' ,force_ascii=False)
#rw_df.to_json(path_or_buf='/Data_file')
```
